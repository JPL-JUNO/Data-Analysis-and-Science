# Coding challenge

```python
import pandas as pd
```

```python
cars = pd.read_csv('used_cars.csv')
cars.head()
```

```python
min_wage = pd.read_csv('minimum_wage.csv')
min_wage.head()
```

```python
cars.pivot_table(index='Fuel', values='Price', aggfunc='sum')
```

```python
cars.pivot_table(
    index='Manufacturer', columns='Transmission', values='Price', aggfunc='count', margins=True
)
```

```python
cars.pivot_table(
    index=['Year', 'Fuel'], columns='Transmission', values='Price', aggfunc='mean'
)
```

```python
cars.pivot_table(
    index=['Year', 'Fuel'], columns='Transmission', values='Price', aggfunc='mean'
).stack()
```

```python
min_wage.melt(
    id_vars='State', var_name='Year', value_name='Wage'
)
```
