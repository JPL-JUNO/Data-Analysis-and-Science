# Filtering a DataFrame

This chapter covers
1. Reducing a DataFrame’s memory use
2. Extracting DataFrame rows by one or more conditions
3. Filtering a DataFrame for rows that include or exclude null values
4. Selecting column values that fall between a range

```python
import pandas as pd
```

## Optimizing a data set for memory use

```python
pd.read_csv('employees.csv')
```

```python
pd.read_csv('employees.csv', parse_dates=['Start Date'], date_format='%m/%d/%y').head()
```

```python
employees = pd.read_csv('employees.csv', parse_dates=['Start Date'], date_format='%m/%d/%y')
```

```python
employees.info()
```

### Converting data types with the astype method

```python
employees['Mgmt'].astype(bool)
```

```python
employees['Mgmt'] = employees['Mgmt'].astype(bool)
employees.tail()
```

```python
employees.info()
```

```python
employees['Salary'].astype(int)
```

```python
employees['Salary'].fillna(0).astype(int).tail()
```

```python
employees['Salary'] = employees['Salary'].fillna(0).astype(int)
```

```python
employees.info()
```

```python
employees.nunique()
```

```python
employees['Gender'].astype('category')
```

```python
employees['Gender'] = employees['Gender'].astype('category')
```

```python
employees.info()
```

```python
employees['Team'] = employees['Team'].astype('category')
```

```python
employees.info()
```

## Filtering by a single condition

```python
employees['First Name'] == 'Maria'
```

```python
employees[employees['First Name'] == 'Maria']
```

```python
employees[employees['Team'] != 'Finance']
```

```python
employees[employees['Mgmt']]
```

```python
high_earners = employees['Salary'] > 100_000
high_earners
```

```python
employees[high_earners].head()
```

## Filtering by multiple conditions

### The and condition

```python
is_female = employees['Gender'] == 'Female'
is_biz_dev = employees['Team'] == 'Business Dev'
employees[is_female & is_biz_dev].head()
```

### The OR condition

```python
earning_below_40k = employees['Salary'] < 40_000
start_after_2015 = employees['Start Date'] > '2015-01-01'
employees[earning_below_40k | start_after_2015].tail()
```

### Inversion with ~

```python
employees[employees['Salary'] < 100_000].head()
```

```python
employees[~(employees['Salary'] >= 100_000)].head()
```

### Methods for Booleans

```python
employees['Team'].ne('Marketing')
```

## Filtering by condition

### The isin method

```python
all_star_teams = ['Sales', 'Legal', 'Marketing']
mask = employees['Team'].isin(all_star_teams)
employees[mask].head()
```

### The between method

```python
mask = employees['Salary'].between(80_000, 90_000)
employees[mask].head()
```

```python
mask = employees['Start Date'].between('1980-01-01', '1990-01-01')
employees[mask].head()
```

```python
mask = employees['First Name'].between('R', 'S')
employees[mask].head()
```

### The isnull and notnull methods

```python
employees['Team'].isnull().head()
```

```python
employees['Start Date'].isnull().head()
```

```python
employees['Team'].notnull().head()
```

```python
mask = employees['Team'].isnull()
employees[mask].head()
```

```python
mask = employees['First Name'].notnull()
employees[mask].head()
```

### Dealing with null values

```python
employees = pd.read_csv('employees.csv', parse_dates=['Start Date'], date_format='%m/%d/%y')
```

```python
employees.dropna()
```

```python
employees.dropna(how='all')
```

```python
employees.dropna(subset=['Gender']).tail()
```

```python
employees.dropna(subset=['Start Date', 'Salary']).head()
```

```python
employees.dropna(thresh=4).head()
```

## Dealing with duplicates

### The duplicated method

```python
employees['Team'].head()
```

```python
employees['Team'].duplicated().head()
```

```python
employees['Team'].duplicated(keep='first').head()
```

```python
employees['Team'].duplicated(keep='last')
```

```python
~employees['Team'].duplicated().head()
```

```python
mask = ~employees['Team'].duplicated()
employees[mask]
```

### The drop_duplicates method

```python
employees.drop_duplicates()
```

```python
employees.drop_duplicates(subset=['Team'])
```

```python
employees.drop_duplicates(subset=['Team'], keep='last')
```

```python
employees.drop_duplicates(subset=['First Name'], keep=False)
```

```python
name_mask = employees['First Name'] == 'Douglas'
gender_mask = employees['Gender'] == 'Male'
employees[name_mask & gender_mask]
```

```python
employees.drop_duplicates(subset=['Team', 'Gender']).head()
```
