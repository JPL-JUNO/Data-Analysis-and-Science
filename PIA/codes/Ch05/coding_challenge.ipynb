```python
import pandas as pd
```

```python
pd.read_csv('netflix.csv')
```

```python
netflix = pd.read_csv('netflix.csv', parse_dates=['date_added'], date_format='%d-%b-%y')
netflix
```

```python
netflix.info()
```

```python
netflix['type'] = netflix['type'].astype('category')
```

```python
netflix.info()
```

```python
mask = netflix['title'] == 'Limitless'
netflix[mask]
```

```python
director_mask = netflix['director'] == 'Robert Rodriguez'
type_mask = netflix['type'] == 'Movie'
netflix[director_mask & type_mask]
```

```python
date_added_mask = netflix['date_added'] == '2019-07-31'
director_mask = netflix['director'] == 'Robert Altman'
netflix[date_added_mask | director_mask]
```

```python
directors = ['Orson Welles', 'Aditya Kripalani', 'Sam Raimi']
directors_mask = netflix['director'].isin(directors)
netflix[directors_mask]
```

```python
date_added_mask = netflix['date_added'].between('2019-05-01', '2019-06-01')
netflix[date_added_mask].head()
```

```python
netflix.dropna(subset=['director']).head()
```

```python
netflix.drop_duplicates(subset=['date_added'], keep=False)
# date_added出现重复的将被删除，并且一条也不保留
```
