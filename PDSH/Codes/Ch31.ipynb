# Multiple Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

## plt.axes: Subplots by Hand

```python
ax1 = plt.axes()
ax2 = plt.axes([.65, .65, .2, .2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([.1, .5, .8, .4], 
                   xticklabels=[], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([.1, .1, .8, .4], ylim=(-1.2, 1.2))
```

## ``plt.subplot``: Simple Grids of Subplots

```python
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(.5, .5, str((2, 3, i)),
             fontsize=18, ha='center')
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=.4, wspace=.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(.5, .5, str((2, 3, i)),
            fontsize=18, ha='center')
```

## ``plt.subplots``: The Whole Grid in One Go

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
```

```python
for i in range(2):
    for j in range(3):
        ax[i, j].text(.5, .5, str((i, j)),
                      fontsize=18, ha='center')
fig
```

## plt.GridSpec: More Complicated Arrangements

```python
grid = plt.GridSpec(2, 3, wspace=.4, hspace=.3)
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2])
plt.show()
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
rng = np.random.default_rng(seed=1701)
x, y = rng.multivariate_normal(mean, cov, 3000).T

fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=.4, wspace=.4)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

main_ax.plot(x, y, 'ok', markersize=3, alpha=.2)
x_hist.hist(x, bins=40, histtype='stepfilled', orientation='vertical', color='gray')
x_hist.invert_yaxis()
y_hist.hist(y, bins=40, histtype='stepfilled', orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```
