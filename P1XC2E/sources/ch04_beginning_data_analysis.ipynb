## 通过更改数据类型来减少内

pandas 对许多数据类型都有精确的技术定义。但是，当你从 CSV 等无类型格式加载数据时，pandas 必须推断类型。

```python
different_cols = [
    'RELAFFIL',
    'SATMTMID',
    'CURROPER',
    'INSTNM',
    'STABBR'
]
```

```python
col2 = college.loc[:, different_cols]
col2.head()
```

```python
col2.dtypes
```

```python
original_mem = col2.memory_usage(deep=True)
original_mem
```

```python
col2['RELAFFIL'] = col2['RELAFFIL'].astype(np.int8)
```

```python
col2.dtypes
```

```python
col2.memory_usage(deep=True)
```

为了节省更多内存，如果 object 数据类型的基数（唯一值的数量）相当低，则需要考虑将其更改为分类。让我们首先检查两个 object 列的唯一值的数量：

```python
col2.select_dtypes(include=['object']).nunique()
```

```python
col2['STABBR'] = col2['STABBR'].astype('category')
col2.dtypes
```

```python
new_mem = col2.memory_usage(deep=True)
new_mem
```

```python
new_mem/original_mem
```

对象数据类型列中每个值的内存是不一致的。与其他数据类型不同，每个值没有预定义的内存量。为了让 pandas 提取对象数据类型列的准确内存量，必须在 `.memory_usage` 方法中将 `deep` 参数设置为True 。

## 选择最大的和的最小的

```python
movie = pd.read_csv('data/movie.csv')
movie2 = movie[["movie_title", "imdb_score", "budget"]]
movie2
```

```python
movie2.nlargest(100, 'imdb_score').head()
```

```python
movie2.nlargest(100, 'imdb_score').nsmallest(5, 'budget')
```

`.nlargest` 方法的第一个参数n 必须是整数，用于选择要返回的行数。第二个参数 `columns` 以字符串形式接受列名。

## 通过排序选择每组中最大的一个

数据分析过程中最基本、最常见的操作之一是选择包含组中某列最大值的行。例如，这就像查找每年收视率最高的电影或按内容评级计算票房最高的电影。

```python
movie[['movie_title', 'title_year', 'imdb_score']].sort_values('title_year', ascending=True)
```

```python
movie[['movie_title', 'title_year', 'imdb_score']].sort_values(['title_year', 'imdb_score'], ascending=True)
```

```python
movie[['movie_title', 'title_year', 'imdb_score']].sort_values(['title_year', 'imdb_score'], ascending=False).drop_duplicates(subset='title_year')
```

```python
movie[['movie_title', 'title_year', 'imdb_score']].groupby('title_year', as_index=False).apply(
    lambda df:df.sort_values('imdb_score', ascending=False).head(1)
).droplevel(0).sort_values('title_year', ascending=False)
```

```python
(
    movie[['movie_title', 'title_year', 'imdb_score']]
    .sort_values(['title_year', 'imdb_score'], ascending=False)
    .drop_duplicates(subset='title_year')
)
```

默认情况下，`.drop_duplicates`会保留值的第一次出现，但可以通过传递 `keep='last'` 来选择每个组的最后一行或传递 `keep=False` 来完全删除所有重复项，从而修改此行为。

## 使用 `sort_values` 复制 `nlargest`

```python
(
    movie[['movie_title', 'imdb_score', 'budget']]
    .nlargest(100, 'imdb_score')
    .nsmallest(5, 'budget')
)
# 评分前 100 中，预算最低的 5 个
```

```python
(
    movie[['movie_title', 'budget', 'imdb_score']]
    .sort_values('imdb_score', ascending=False)
    .head(100)
    .sort_values('budget').head(5)
)
```

```python
(
    movie[['movie_title', 'budget', 'imdb_score']]
    .sort_values('imdb_score', ascending=False, kind='mergesort')
    .head(100)
    .sort_values('budget').head(5)
)
```
