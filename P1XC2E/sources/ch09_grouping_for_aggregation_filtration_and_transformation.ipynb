## 分组后删除 `MultiIndex`

当使用 `groupby` 时，不可避免地会创建一个 `MultiIndex`。多重索引可以发生在索引和列中。具有多索引的 DataFrame 更难以定位，并且偶尔也会出现令人困惑的列名称。

这里，我们使用 .groupby 方法执行聚合，以创建一个具有行和列 `MultiIndex` 的 DataFrame。然后，我们操作索引，使其具有单一级别并且列名具有描述性。

## 自定义聚合函数

pandas 提供了许多与 groupby 对象一起使用的聚合函数。在某些时候，您可能需要编写自己的自定义用户定义函数，但 pandas 或 NumPy 中不存在该函数。

```python
college = pd.read_csv("data/college.csv")
college.groupby("STABBR")["UGDS"].agg(["mean", "std"]).round(0)
```

```python
def max_deviation(s):
    std_score = (s-s.mean())/s.std()
    return std_score.abs().max()
```

```python
(
    college.groupby('STABBR')
    ['UGDS']
    .agg(max_deviation)
    .round(1)
)
```

```python
(
    college.groupby('STABBR')
    [['UGDS', 'SATVRMID', 'SATMTMID']]
    .agg(max_deviation)
    .round(1).head()
)
```

还可以将自定义聚合函数与预建函数一起使用。

```python
(
    college.groupby('STABBR')
    [['UGDS', 'SATVRMID', 'SATMTMID']]
    .agg([max_deviation, 'mean', 'std'])
    .round(1).head()
)
```

请注意，pandas 使用函数的名称作为返回列的名称。您可以使用 `.rename` 方法直接更改列名，也可以修改函数属性 `.__name__`：

## 使用 `*args` 和 `**kwargs` 自定义聚合函数

在编写您自己的用户定义的自定义聚合函数时，**pandas 会隐式地将每个聚合列作为 Series 一次传递给它**。有时，您需要向函数传递比 Series 本身更多的参数。

`.agg` 的签名是 `agg(func, *args, **kwargs)`。`func` 参数是一个 Reduced 函数、Reducing 方法的字符串名称、Reducing 函数列表或将列映射到函数或函数列表的字典。此外，正如我们所见，您可以使用关键字参数来创建命名聚合。

如果您有一个需要使用其他参数的归约函数，则可以利用 `*args` 和 `**kwargs` 参数将参数传递给归约函数。您可以使用 `*args` 将任意数量的位置参数传递给您的自定义聚合函数。同样，`**kwargs` 允许您传递任意数量的关键字参数。

```python
def pct_between_1_3k(s):
    return (
        s.between(1000, 3000).mean()*100
    )
```

```python
(
    college.groupby(['STABBR', 'RELAFFIL'])
    ['UGDS']
    .agg(pct_between_1_3k)
    .round(1)
)
```

```python
def pct_between(s, low, high):
    return s.between(low, high).mean() * 100

(
    college.groupby(['STABBR', 'RELAFFIL'])
    ['UGDS']
    .agg(pct_between, 1_000, 3_000)
    .round(1)
)
```

```python
(
    college.groupby(['STABBR', 'RELAFFIL'])
    ['UGDS']
    .agg(pct_between, high=10_000, low=3_000)
    .round(1)
)
```

如果我们想要调用多个聚合函数，并且其中一些需要参数，我们可以利用 Python 的闭包功能来创建一个新函数，该函数在其调用环境中封闭参数：

```python
def between_n_m(n, m):
    def wrapper(ser):
        return pct_between(ser, n, m)
    
    wrapper.__name__ = f'between_{n}_{m}'
    return wrapper

(
    college.groupby(['STABBR', 'RELAFFIL'])
    ['UGDS']
    .agg([between_n_m(1000, 10000), 'max', 'mean'])
    .round(1)
)
# 一层一层调用，ser 最后一层才传入？
```

## 筛选少数族裔占多数的州

当使用.groupby方法时，我们可以过滤掉组。groupby对象的.filter方法接受一个函数，该函数必须返回True或False来指示是否保留某个组。

需要注意的是，当应用.filter方法时，结果不会使用分组列作为索引，而是保留原始索引。DataFrame .filter方法过滤列，而不是值。

```python
grouped = college.groupby('STABBR')
grouped.ngroups
```

```python
college['STABBR'].nunique()
```

分组变量有一个 `.filter` 方法，它接受自定义函数确定是否保留某个组。自定义函数接受当前组的 DataFrame 并需要返回布尔值。

```python
def check_minority(df, threshold):
    minority_pct = 1 - df['UGDS_WHITE']
    total_minority = (df['UGDS'] * minority_pct).sum()
    
    total_ugds = df['UGDS'].sum()
    total_minority_pct = total_minority / total_ugds
    
    return total_minority_pct > threshold
```

```python
college_filtered = grouped.filter(check_minority, threshold=.5)
college_filtered
```

## 使用 apply 计算每个州的加权平均 SAT 分数

groupby 对象有四种方法，它们接受一个或多个函数来对每个组执行计算。这四种方法是 `.agg`、`.filter`、`.transform` 和 `.apply`。前三种方法都有函数必须返回的非常具体的输出。`.agg` 必须返回标量值，`.filter`必须返回布尔值，`.transform` 必须返回与传递的组长度相同的 Series 或 DataFrame。

但是，`.apply` 方法可以返回标量值、Series 甚至任何形状的 DataFrame，因此非常灵活。它每个组只调用一次（在 DataFrame上），而 `.transform` 和 `.agg` 方法每个聚合列调用一次（在 Series 上）。`.apply` 方法在同时对多个列进行操作时能够返回单个对象，这使得此配方中的计算成为可能。

```python
subset = ['UGDS', 'SATMTMID', 'SATVRMID']
college2 = college.dropna(subset=subset)
college2.shape
```

```python
def weighted_math_average(df):
    weighted_math = df['UGDS'] * df['SATMTMID']
    return int(weighted_math.sum() / df['UGDS'].sum())
```

```python
college2.groupby('STABBR').apply(weighted_math_average)
```

`.apply` 的一个不错的功能是，您可以通过返回一个 Series 来创建多个新 columns。此返回 Series 的索引将是新列的名称。

```python
def weighted_average(df):
    weight_m = df['UGDS'] * df['SATMTMID']
    weight_v = df['UGDS'] * df['SATVRMID']
    wm_avg = weight_m.sum() / df['UGDS'].sum()
    wv_avg = weight_v.sum() / df['UGDS'].sum()
    
    data = {
        'w_math_avg': wv_avg,
        'v_verbal_avg': wv_avg,
        'verbal_avg': df['SATMTMID'].mean(),
        'count': len(df)
    }
    return pd.Series(data)
```

```python
(
    college2.groupby(
        'STABBR'
    ).apply(weighted_average)
    .astype(int)
    .head()
)
```

```python
from scipy.stats import gmean, hmean

def calculate_means(df):
    df_means = pd.DataFrame(index=['Arithmetic', 'Weighted',
                                   'Geometric', 'Harmonic'])
    cols = ["SATMTMID", 'SATVRMID']
    
    for col in cols:
        arithmetic = df[col].mean()
        weighted = np.average(df[col], weights=df['UGDS'])
        
        geometric = gmean(df[col])
        harmonic = hmean(df[col])
        
        df_means[col] = [arithmetic, weighted, geometric, harmonic]
        
    df_means['count'] = len(df)
    
    return df_means.astype(int)
```

```python
(
    college2
    .groupby('STABBR')
    .apply(calculate_means)
)
```

## 按连续变量分组

在 Pandas 中分组时，通常使用具有离散重复值的列。如果没有重复值，则分组将毫无意义，因为每个组只有一行。连续数字列通常具有很少的重复值，并且通常不用于形成组。但是，如果我们可以通过将每个值放入一个容器、对其进行舍入或使用其他映射将具有连续值的列转换为离散列，那么对它们进行分组就很有意义了。

```python
flights = pd.read_csv('data/flights.csv')
flights.sample()
```

```python
def max_streak(s):
    s1 = s.cumsum()
    return (
        s.mul(s1)
        .diff()
        .where(lambda x : x < 0)
        .ffill()
        .add(s1, fill_value=0)
        .max()
    )
```

```python
(
    flights.assign(
        ON_TIME=flights['ARR_DELAY'].lt(15).astype(int)
        )
        .sort_values(['MONTH', 'DAY', 'SCHED_DEP'])
        .groupby(['AIRLINE', 'ORG_AIR'])
        ['ON_TIME']
        .agg(['mean', 'size', max_streak])
        .round(2)
)
```

```python
def max_delay_streak(df):
    df = df.reset_index(drop=True)
    late = 1 - df['ON_TIME']
    late_sum = late.cumsum()
    streak = (late
        .mul(late_sum)
        .diff()
        .where(lambda x: x < 0) 
        .ffill()
        .add(late_sum, fill_value=0)
    )
    last_idx = streak.idxmax()
    first_idx = last_idx - streak.max() + 1
    if first_idx > last_idx:
        res = (
            df.loc[[last_idx, last_idx], ['MONTH', 'DAY']]
            .assign(streak=streak.max())
        )
    else:
        res = (df
        .loc[[first_idx, last_idx], ['MONTH', 'DAY']]
        .assign(streak=streak.max())
    )
    res.index = ['first', 'last']
    return res
```

```python
(flights
    .assign(ON_TIME=flights['ARR_DELAY'].lt(15).astype(int))
    .sort_values(['MONTH', 'DAY', 'SCHED_DEP']) 
    .groupby(['AIRLINE', 'ORG_AIR'])
    .apply(max_delay_streak)
    .sort_values('streak', ascending=False)
)
```
