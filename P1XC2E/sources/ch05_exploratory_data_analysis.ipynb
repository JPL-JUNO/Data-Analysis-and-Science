查找有问题的字符的另一种方法是使用带有正则表达式的 `.str.extract` 方法：

```python
fueleco.rangeA.value_counts()
```

```python
(
    fueleco.rangeA.str.extract(r'([^0-9.])')
    .dropna()
    .apply(lambda row: ''.join(row), axis=1)
    .value_counts()
)
```

```python
(
    fueleco.rangeA.fillna('0')
    .str.replace('-', '/')
    .str.split('/', expand=True)
    .astype(float)
    .mean(axis=1)
)
```

我们还可以通过对数字列进行分箱处理，将其视为类别。pandas 中有两个强大的函数可以帮助进行分箱处理，即 `cut` 和 `qcut`。我们可以使用 `cut` **将其切分为等宽的箱**，或者指定箱宽（(max-min)/n）。对于 rangeA 列，大多数值都是空的，我们将其替换为 0，因此 10 个等宽箱如下所示：

```python
(
    fueleco.rangeA.fillna('0')
    .str.replace('-', '/')
    .str.split('/', expand=True)
    .astype(float)
    .mean(axis=1)
    .pipe(lambda ser_:pd.cut(ser_, 10))
    .value_counts()
)
```

或者，`qcut`（分位数切割）会将条目切割成大小相同的容器。由于rangeA列严重倾斜，并且大多数条目为 0，我们无法将 0 量化为多个容器，因此它失败了。

```python
try:
    (
    fueleco.rangeA.fillna('0')
    .str.replace('-', '/')
    .str.split('/', expand=True)
    .astype(float)
    .mean(axis=1)
    .pipe(lambda ser_:pd.qcut(ser_, 10))
    .value_counts()
)
except ValueError as e:
    print(e)
```
