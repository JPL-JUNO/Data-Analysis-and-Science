我们还可以通过对数字列进行分箱处理，将其视为类别。pandas 中有两个强大的函数可以帮助进行分箱处理，即 `cut` 和 `qcut`。我们可以使用 `cut` **将其切分为等宽的箱**，或者指定箱宽（(max-min)/n）。对于 rangeA 列，大多数值都是空的，我们将其替换为 0，因此 10 个等宽箱如下所示：

```python
(
    fueleco.rangeA.fillna('0')
    .str.replace('-', '/')
    .str.split('/', expand=True)
    .astype(float)
    .mean(axis=1)
    .pipe(lambda ser_:pd.cut(ser_, 10))
    .value_counts()
)
```

或者，`qcut`（分位数切割）会将条目切割成大小相同的容器。由于rangeA列严重倾斜，并且大多数条目为 0，我们无法将 0 量化为多个容器，因此它失败了。

```python
try:
    (
    fueleco.rangeA.fillna('0')
    .str.replace('-', '/')
    .str.split('/', expand=True)
    .astype(float)
    .mean(axis=1)
    .pipe(lambda ser_:pd.qcut(ser_, 10))
    .value_counts()
)
except ValueError as e:
    print(e)
```

## 将分类值与分类值进行比较

```python
def generalize(ser, match_name, default):
    seen = None
    for match, name in match_name:
        mask = ser.str.contains(match)
        if seen is None:
            seen = mask
        else:
            seen |= mask
        # Replace values where the condition is False.
        ser = ser.where(~mask, name)
    ser = ser.where(seen, default)
    return ser
```

```python
makes = ["Ford", "Tesla", "BMW", "Toyota"]
```

```python
data = fueleco[fueleco.make.isin(makes)].assign(
    SClass=lambda df_: generalize(df_.VClass, [
        ('Seaters', "Car"),
        ('Car', 'Car'),
        ('Utility', 'SUV'),
        ('Truck', 'Truck'),
        ('Van', 'Van'),
        ('van', 'Van'),
        ('Wagon', 'Wagon')
    ], 'other')
)
```

使用 [Cramér's V](https://stackoverflow.com/questions/46498455/categorical-features-correlation/46498792#46498792) 测量表示类别相关性

```python
import scipy.stats as ss
import numpy as np

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k=confusion_matrix.shape
    phi2corr = max(
        0,phi2-((k-1)*(r-1))/(n-1)
    )
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return np.sqrt(
        phi2corr/min((kcorr-1), (rcorr-1))
    )
```

```python
cramers_v(data.make, data.SClass)
```

```python
data.make.corr(data.SClass, cramers_v)
```

通过对交叉表进行归一化，可视化各组的相对大小，并制作堆积条形图：

```python
fig, ax = plt.subplots(figsize=(10, 8))
(
    data.pipe(
        lambda df_: pd.crosstab(df_.make, df_.SClass)
    )
    .pipe(lambda df_: df_.div(df_.sum(axis=1), axis=0))
    .plot.bar(stacked=True, ax=ax)
)
fig.savefig('img/c5-barstacked.png', dpi=300, bbox_inches='tight')
```
