使用 `concat()` 将数据帧堆叠（即连接）在一起。所有要连接的数据框都以列表形式传递。`concat()` 盲目地将数据帧堆叠在一起。

```python
row_concat = pd.concat([df1, df2, df3])
row_concat
```

```python
row_concat.iloc[3:]
```

```python
row_concat.loc[3]
```

```python
new_row_series = pd.Series(['n1', 'n2', 'n3', 'n4'])
# new_row_series = pd.Series(['n1', 'n2', 'n3', 'n4'], name='A')
new_row_series
```

```python
pd.concat([df1, new_row_series])
```

因为这里的 Series 没有 `name` 属性，并且 `df1` 中也找不到对应的列名，因此被单独作为一列存储，同时是增加行，所以就导致很多没有对应的缺失值（`NaN`）

```python
pd.concat([df1, new_row_series], axis=1)
```

```python
new_row_df = pd.DataFrame(
    data=[['n1', 'n2', 'n3', 'n4']],
    columns=df1.columns
)
new_row_df
```

```python
pd.concat([df1, new_row_df])
```

#### 忽略原数据的索引

###  Add Columns

```python
col_concat = pd.concat([df1, df2, df3], axis='columns')
col_concat
```

```python
col_concat['A']
```

```python
pd.concat([df1, df2, df3], axis='columns', ignore_index=True)
```

###  Concatenate with Different Indices

```python
df1.columns = ['A', 'B', 'C', 'D']
df2.columns = ['E', 'F', 'G', 'H']
df3.columns = ['A', 'C', 'F', 'H']
df1
```

```python
df2
```

```python
df3
```

```python
row_concat = pd.concat([df1, df2, df3])
row_concat
```

这是因为列名上都对不齐，采用的并集操作。

```python
pd.concat([df1, df2, df3], join='inner')
```

```python
pd.concat([df1, df3], ignore_index=True, join='inner')
# 只有字段的并集才会被连接
```

#### 连接具有不同行的列

```python
df1.index = [0, 1, 2, 3]
df2.index = [4, 5, 6, 7]
df3.index = [0, 2, 5, 7]
df1
```

```python
df2
```

```python
df3
```

```python
col_concat = pd.concat([df1, df2, df3], axis='columns')
col_concat
```

```python
pd.concat([df1, df3], axis='columns', join='inner')
# 只有索引的并集才会被连接
```
