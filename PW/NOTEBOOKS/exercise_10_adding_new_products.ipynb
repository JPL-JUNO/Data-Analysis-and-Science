```python
import pandas as pd
import numpy as np
```

```python
new_products = pd.DataFrame(
    [
        {
            "product_id": 24,
            "name": "phone",
            "wholesale_price": 200,
            "retail_price": 500,
        },
        {"product_id": 16, "name": "apple", "wholesale_price": 0.5, "retail_price": 1},
        {"product_id": 17, "name": "pear", "wholesale_price": 0.6, "retail_price": 1.2},
    ],
    index=range(5, 8),
)

df = pd.DataFrame(
    [
        {
            "product_id": 23,
            "name": "computer",
            "wholesale_price": 500,
            "retail_price": 1000,
            "sales": 100,
        },
        {
            "product_id": 96,
            "name": "Python Workout",
            "wholesale_price": 35,
            "retail_price": 75,
            "sales": 1000,
        },
        {
            "product_id": 97,
            "name": "Pandas Workout",
            "wholesale_price": 35,
            "retail_price": 75,
            "sales": 500,
        },
        {
            "product_id": 15,
            "name": "banana",
            "wholesale_price": 0.5,
            "retail_price": 1,
            "sales": 200,
        },
        {
            "product_id": 87,
            "name": "sandwich",
            "wholesale_price": 3,
            "retail_price": 5,
            "sales": 300,
        },
    ]
)
```

```python
df
```

```python
new_products
```

```python
df = pd.concat([df, new_products])
df
```

每当 pandas 想要使用 `NaN` 时，它都需要将列设置为具有浮点数据类型。

> 在 NumPy 中，将浮点值分配给具有整数 dtype 的数组会导致浮点被无声地截断。而尝试将 NaN（虽然是一个奇怪的浮点）分配给具有整数 dtype 的数组会导致错误，NumPy 会指示 NaN 没有整数值。相比之下，Pandas 会尝试适应，将 dtype 更改为 float64 以适应 NaN 值。但它不会就此发出警告！不会丢失数据，但可能会对未明确要求的 dtype 更改感到惊讶。

```python
df.loc[[5, 6, 7], 'sales'] = [100, 200, 75]
df
```

```python
((df.retail_price - df.wholesale_price) * df.sales).sum()
```

```python
df.loc[8] = [99, 'persimmon', 2, 4.5, 1]
df
```

```python
df['department'] = ['electronics', 'books', 'books', 'food', 'food',
                    'electronics', 'food', 'food', 'food']
df
```

```python
df['current_net'] = (df.retail_price - df.wholesale_price) * df.sales

df.current_net[df.department == 'electronics'].describe()
```

```python
df.query('department == "food"')['current_net'].describe()
```

### Getting answers with the query method

```python
df[df['retail_price'] > 300]
```

```python
df.query('retail_price > 300 & sales % 2 == 0')
```

> 请注意，`query` 不能用在赋值的左侧。对于较小的数据帧，查询不仅会过度，还会减慢代码速度。但是，当处理超过 10,000 行的数据帧时，查询的速度会比传统的查询方式快得多。此外，它可以使用远更少的内存。
