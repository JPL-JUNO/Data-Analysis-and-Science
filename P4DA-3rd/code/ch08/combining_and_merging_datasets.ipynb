# Combining and Merging Datasets

```python
import pandas as pd
```

## Database-Style DataFrame Joins

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': pd.Series(range(7), dtype='Int64')})
df1
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': pd.Series(range(3), dtype='Int64')})
df2
```

```python
pd.merge(df1, df2)
# Note that I didn’t specify which column to join on. If that information is not
# specified, pandas.merge uses the overlapping column names as the keys. It’s a good
# practice to specify explicitly, though:
```

```python
pd.merge(df1, df2, on='key')
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': pd.Series(range(7), dtype='Int64')})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': pd.Series(range(3), dtype='Int64')})
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey', how='outer')
```

```python
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "b"],
                    "data1": pd.Series(range(6), dtype="Int64")})
df2 = pd.DataFrame({"key": ["a", "b", "a", "b", "d"],
                    "data2": pd.Series(range(5), dtype="Int64")})
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='key', how='left')
```

```python
pd.merge(df1, df2, how='inner')
```

```python
left = pd.DataFrame({"key1": ["foo", "foo", "bar"],
                     "key2": ["one", "two", "one"],
                     "lval": pd.Series([1, 2, 3], dtype='Int64')})
right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"],
                      "key2": ["one", "one", "one", "two"],
                      "rval": pd.Series([4, 5, 6, 7], dtype='Int64')})
left
```

```python
right
```

```python
# To merge with multiple keys, pass a list of column names:
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
pd.merge(left, right, on='key1')
```

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```

## Merging on Index

```python
left1 = pd.DataFrame({"key": ["a", "b", "a", "a", "b", "c"],
                      "value": pd.Series(range(6), dtype="Int64")})
right1 = pd.DataFrame({"group_val": [3.5, 7]}, index=["a", "b"])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
lefth = pd.DataFrame({"key1": ["Ohio", "Ohio", "Ohio", "Nevada", "Nevada"],
                      "key2": [2000, 2001, 2002, 2001, 2002],
                      "data": pd.Series(range(5), dtype="Int64")})
lefth
```

```python
righth_index = pd.MultiIndex.from_arrays([
    ["Nevada", "Nevada", "Ohio", "Ohio", "Ohio", "Ohio"],
    [2001, 2000, 2000, 2000, 2001, 2002]]
)
righth = pd.DataFrame({"event1": pd.Series([0, 2, 4, 6, 8, 10], dtype="Int64", index=righth_index),
                       "event2": pd.Series([1, 3, 5, 7, 9, 11], dtype="Int64", index=righth_index)})
righth
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True)
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True, how='outer')
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                     index=["a", "c", "e"],
                     columns=["Ohio", "Nevada"]).astype("Int64")
left2
```

```python
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=["b", "c", "d", "e"],
                      columns=["Missouri", "Alabama"]).astype("Int64")
right2
```

```python
pd.merge(left2, right2, left_index=True, right_index=True, how='outer')
```

```python
# 简写全部由索引 merge 的代码，而且 columns 不能 overlapping
left2.join(right2, how='outer')
```

```python
left2.join(right2)
# use left join by default
# You can think this method as joining data into the object whose join method was called 
```

```python
another = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
                       index=["a", "c", "e", "f"],
                       columns=["New York", "Oregon"])
another
```

```python
left2.join([right2, another])
```

```python
left2.join([right2, another], how='outer')
```

## Concatenating Along an Axis

```python
import numpy as np
```

```python
arr = np.arange(12).reshape((3, 4))
arr
```

```python
np.concatenate([arr, arr], axis=1)
```

```python
s1 =  pd.Series([0, 1], index=['a', 'b'], dtype='Int64')
s1
```

```python
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'], dtype='Int64')
s2
```

```python
s3 = pd.Series([5, 6], index=['f', 'g'], dtype='Int64')
s3
```

```python
pd.concat([s1, s2, s3])
```

```python
pd.concat([s1, s2, s3], axis='columns')
```

```python
s4 = pd.concat([s1, s3])
s4
```

```python
pd.concat([s1, s4], axis='columns')
```

```python
pd.concat([s1, s4], axis='columns', join='inner')
```

```python
result = pd.concat([s1, s1, s3], keys=['one', 'two', 'three'])
result
```

```python
result.unstack()
```

```python
pd.concat([s1, s2, s3], axis='columns', keys=['one', 'two', 'three'])
```

```python
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=['a', 'b', 'c'],
                   columns=['one', 'two'])
df1
```

```python
df2 = pd.DataFrame(np.arange(4).reshape(2, 2) + 5, index=['a', 'c'], columns=['three', 'four'])
df2
```

```python
pd.concat([df1, df2], axis='columns', keys=['level1', 'level2'])
```

```python
pd.concat({'key1': df1, 'key2': df2}, axis='columns')
```

```python
pd.concat([df1, df2], axis='columns', keys=['level1', 'level2'], names=['upper', 'lower'])
```

```python
df1 = pd.DataFrame(np.random.standard_normal((3, 4)), columns=['a', 'b', 'c', 'd'])
df1
```

```python
df2 = pd.DataFrame(np.random.standard_normal((2, 3)), columns=['b', 'd', 'a'])
df2
```

```python
pd.concat([df1, df2], ignore_index=True)
```

## Combining Data with Overlap

```python
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],
              index=['f', 'e', 'd', 'c', 'b', 'a'])
a
```

```python
b = pd.Series(np.arange(len(a), dtype=np.float64),
              index=['a', 'b', 'c', 'd', 'e', 'f'])
b
```

```python
np.where(pd.isna(a), b, a)
```

```python
a.combine_first(b)
```

```python
df1 = pd.DataFrame({'a': [1, np.nan, 5., np.nan],
                    'b': [np.nan, 2., np.nan, 6.],
                    'c': range(2, 18, 4)})
df1
```

```python
df2 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],
                    'b': [np.nan, 3., 4., 6. ,8.]})
df2
```

```python
df1.combine_first(df2)
```
