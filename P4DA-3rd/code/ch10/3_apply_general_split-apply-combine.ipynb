# apply：一般性的“拆分－应用－合并”

```python
import pandas as pd
import numpy as np
```

```python
tips = pd.read_csv('../examples/tips.csv')
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

```python
def top(df, n=5, column='tip_pct'):
    return df.sort_values(column, ascending=False)[:n]
top(tips, n=6)
```

```python
tips.groupby('smoker').apply(top)
```

```python
tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')
```

```python
result = tips.groupby('smoker')['tip_pct'].describe()
result
```

```python
result.unstack('smoker')
```

```python
def f(group):
    return group.describe()

grouped = tips.groupby('smoker')['tip_pct']
grouped.apply(f)
```

## 禁止分组键

```python
tips.groupby('smoker', group_keys=False).apply(top)
```

```python
tips.groupby('smoker').apply(top)
```

## 分位数和桶分析

```python
frame = pd.DataFrame({'data1': np.random.standard_normal(1_000),
                      'data2': np.random.standard_normal(1_000)})
frame.head()
```

```python
quantiles = pd.cut(frame['data1'], 4)
quantiles.head(10)
```

```python
def get_stats(group):
    return pd.DataFrame({
        'min': group.min(), 'max': group.max(),
        'count': group.count(), 'mean': group.mean(),
    })
```

```python
grouped = frame.groupby(quantiles, observed=False)
grouped
```

```python
grouped.apply(get_stats)
```

```python
frame.min()
```

```python
# 当然有更加简单的方式
grouped.agg(['min', 'max', 'mean', 'count'])
```

```python
quartiles_samp = pd.qcut(frame['data1'], 4, labels=False)
quartiles_samp
# 传入 labels=False 即可只获取分位数的编号：
```

```python
pd.qcut(frame['data1'], 4)
```

```python
grouped = frame.groupby(quartiles_samp)
grouped.apply(get_stats)
# 可以看到 count 与 pd.cut() 的结果完全不一样
```
