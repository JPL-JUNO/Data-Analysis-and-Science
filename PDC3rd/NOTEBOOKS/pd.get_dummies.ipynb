```python
import pandas as pd
import numpy as np
```

```python
ser = pd.Series([
    "green",
    "brown",
    "blue",
    "amber",
    "hazel",
    "amber",
    "green",
    "blue",
    "green",
], name="eye_colors", dtype=pd.StringDtype())
ser
```

```python
pd.get_dummies(ser)
```

将其作为参数传递给 `pd.get_dummies` 将创建一个类似索引的 `pd.DataFrame`，其中每个都有一个布尔列。每行都有一个 `True` 列，将其映射回其原始值；同一行中的所有其他列将为 False`。

如果我们对默认的列名不满意，可以通过添加前缀来修改它们。数据建模中的一个常见约定是在布尔列前加上 `is_` 前缀：

```python
pd.get_dummies(ser, prefix='is')
```
