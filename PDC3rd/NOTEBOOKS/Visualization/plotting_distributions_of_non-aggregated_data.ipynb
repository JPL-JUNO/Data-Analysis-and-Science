```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.ion()
```

```python
np.random.seed(42)

ser = pd.Series(
    np.random.default_rng().normal(size=10_000),
    dtype=pd.Float64Dtype()
)
ser.head()
```

```python
ser.plot(kind='hist')
```

```python
ser.plot(kind='hist', bins=100)
```

```python
df = pd.DataFrame({
    'normal': np.random.default_rng().normal(size=10_000),
    'triangular': np.random.default_rng().triangular(-2, 0, 2, size=10_000)
})
df = df.convert_dtypes(dtype_backend='numpy_nullable')
df.head()
```

```python
df.plot(kind='hist')
```

```python
df.plot(kind='hist', alpha=.5)
```

```python
df.plot(kind='hist', subplots=True)
```

```python
df.plot(kind='hist', bins=100, alpha=.5)
```

虽然直方图很常用，但不幸的是，分箱的选择会对数据的解释产生影响；不会希望仅仅因为选择了“错误”的分箱数量而改变对数据的解释！

幸运的是，可以使用一种类似但可能更强大的可视化方法，它不需要选择任何类型的分箱策略，即核密度估计（或KDE）图。

```python
ser.plot(kind='kde')
```

```python
df.plot(kind='kde')
```
