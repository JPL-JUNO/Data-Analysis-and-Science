```python
import pandas as pd
import numpy as np
```

Top N 分析是一种常用技术，通过该技术，可以根据数据在用单个变量衡量时的表现来过滤数据。大多数分析工具都能够帮助过滤数据，以回答诸如销售额排名前 10 位的客户是谁？或库存最低的 10 种产品是什么？等问题。将这些内容串联起来，甚至可以形成引人注目的新闻标题，例如在前 100 所大学中，这 5 所大学的学费最低，或在最宜居的 50 个城市中，这 10 个城市的生活成本最低。

```python
df = pd.read_csv('../DATA/movie.csv', usecols=["movie_title", "imdb_score", "budget", "gross"],
                 dtype_backend='numpy_nullable')
df.head()
```

```python
df.nlargest(100, 'imdb_score').head()
```

```python
df.nlargest(100, 'imdb_score').head().nsmallest(5, 'budget')
```

可以将列名列表作为 `pd.DataFrame.nlargest` 和 `pd.DataFrame.nsmallest` 的 `columns`参数传递。这仅在存在重复值共享列表中第一列中第 `n` 个排名位置的情况下才有用。

```python
df.nlargest(10, 'imdb_score')
```

```python
df[df.imdb_score == 8.9]
```

> 这就产生问题，就是评分 8.9 的影片有 5 部，但是仅出现了两个在前十里面，这仅仅因为这两部的 index 在前面。

```python
df.nlargest(10, ['imdb_score', 'gross'])
```
